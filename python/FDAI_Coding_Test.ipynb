```python
dic = {}
name_id = 0
for i in range(0, len(data)):
    temp = data['hashed'][i].split(',')
    for j in range(0, len(temp)):
        if temp[j] not in dic:
            dic[temp[j]] = {'id' : name_id, 'count' : 1}
            name_id = name_id + 1
        else:
            dic[temp[j]] = {'id' : dic[temp[j]]['id'], 'count' : dic[temp[j]]['count']+1}
```

```python
time_temp = []
for i in range(0,4):
    result = int(time1[i]) - int(time2[i])
    if result < 0:
        if i == 1:
            time_temp[0] = time_temp[0] - 1
            result = int(time1[i]) + 60 - int(time2[i])
        elif i == 2:
            time_temp[1] = time_temp[1] - 1
            result = int(time1[i]) + 60 - int(time2[i])
        elif i == 3:
            time_temp[2] = time_temp[2] - 1
            result = int(time1[i]) + 1000 - int(time2[i])
    time_temp.append(result)
    
total_time = time_temp[0] * 3600 + time_temp[1] * 60 + time_temp[2] + time_temp[3] / 1000
total_time

#--------------------------------테스트종료-------------------------------실패 ㅎㅎ..!
```

```python
total_time = 0
for i in range(0, len(data)):
    time1 = data['call_end_time'][i].split(':')
    time1.append(time1[2].split('.')[0])
    time1.append(time1[2].split('.')[1])
    del time1[2]
    
    time2 = data['call_start_time'][i].split(':')
    time2.append(time2[2].split('.')[0])
    time2.append(time2[2].split('.')[1])
    del time2[2]
    
    time_temp = []
    second_time1 = 0
    second_time2 = 0
    
    date1 = int(data['call_end_day'][i])
    date2 = int(data['call_start_day'][i])
    
    date = date1 - date2
    
    date = date * 24 * 3600
    
    second_time1 = int(time1[0]) * 3600 + int(time1[1]) * 60 + int(time1[2]) + int(time1[3]) / 1000
    second_time2 = int(time2[0]) * 3600 + int(time2[1]) * 60 + int(time2[2]) + int(time2[3]) / 1000
    second_time = second_time1 - second_time2 + date
    total_time = second_time + total_time
```
