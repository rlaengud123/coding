```python
import os
import random
```

```python
rate = float(input())
path = r'F:/cmc_data2'
path_list = os.listdir(path)
all_list1 = []
all_list2 = []
for i in range(0, len(path_list)):
    file_dir = path + "/" + path_list[i]
    file_list = os.listdir(file_dir)
    for j in range(0, len(file_list)):
        temp_list = [file_list[j], str(int(path_list[i])+1)]
        if path_list[i] =='0':
            all_list1.append(temp_list)
        else:
            all_list2.append(temp_list)

random.shuffle(all_list1)
random.shuffle(all_list2)

train_list1 = all_list1[:int(len(all_list1)*rate)]
train_list2 = all_list2[:int(len(all_list2)*rate)]
test_list1 = all_list1[int(len(all_list1)*rate):]
test_list2 = all_list2[int(len(all_list2)*rate):]

train_list = train_list1 + train_list2
test_list = test_list1 + test_list2

f = open("F://test_list01.txt", 'w')
for i in range(0, len(test_list)):
    name = "normal"
    if test_list[i][1] == '2':
        name = "abnormal"
    if i == len(test_list)-1:
        data = name + "/" + test_list[i][0]
    else:
        data = name + "/" + test_list[i][0] + "\n"
    f.write(data)
f.close()

f = open("F://train_list01.txt", 'w')
for i in range(0, len(train_list)):
    name = "normal"
    if train_list[i][1] == '2':
        name = "abnormal"
    if i == len(train_list)-1:
        data = name + "/" + train_list[i][0] + " " + train_list[i][1]
    else:
        data = name + "/" + train_list[i][0] + " " + train_list[i][1] + "\n"
    f.write(data)
f.close()
```

```python
rate = float(input())
path = r'F:/cmc_data2'
path_list = os.listdir(path)
all_list1 = []
all_list2 = []
for i in range(0, len(path_list)):
    file_dir = path + "/" + path_list[i]
    file_list = os.listdir(file_dir)
    for j in range(0, len(file_list)):
        temp_list = [file_list[j], path_list[i]]
        if path_list[i] =='0':
            all_list1.append(temp_list)
        else:
            all_list2.append(temp_list)

random.shuffle(all_list1)
random.shuffle(all_list2)

train_list1 = all_list1[:int(len(all_list1)*rate)]
train_list2 = all_list2[:int(len(all_list2)*rate)]
test_list1 = all_list1[int(len(all_list1)*rate):]
test_list2 = all_list2[int(len(all_list2)*rate):]

train_len1_1 = int(len(train_list1)/3)
train_len1_2 = int(len(train_list1)/3*2)
train_len2_1 = int(len(train_list2)/3)
train_len2_2 = int(len(train_list2)/3*2)

train_list_1 = train_list1[:train_len1_1] + train_list2[:train_len2_1]
train_list_2 = train_list1[train_len1_1:train_len1_2] + train_list2[train_len2_1:train_len2_2]
train_list_3 = train_list1[train_len1_2:] + train_list2[train_len2_2:]

test_len1_1 = int((len(test_list1))/3)
test_len1_2 = int((len(test_list1))/3*2)
test_len2_1 = int((len(test_list2))/3)
test_len2_2 = int((len(test_list2))/3*2)

test_list_1 = test_list1[:test_len1_1] + test_list2[:test_len2_1]
test_list_2 = test_list1[test_len1_1:test_len1_2] + test_list2[test_len2_1:test_len2_2]
test_list_3 = test_list1[test_len1_2:] + test_list2[test_len2_2:]

f = open("F://testlist01.txt", 'w')
for i in range(0, len(test_list_1)):
    if i == len(test_list_1)-1:
        data = test_list_1[i][0] + "/" + test_list_1[i][0] + ".avi"
    else:
        data = test_list_1[i][0] + "/" + test_list_1[i][0] + ".avi" + "\n"
    f.write(data)
f.close()

f = open("F://testlist02.txt", 'w')
for i in range(0, len(test_list_2)):
    if i == len(test_list_2)-1:
        data = test_list_2[i][0] + "/" + test_list_2[i][0] + ".avi"
    else:
        data = test_list_2[i][0] + "/" + test_list_2[i][0] + ".avi" + "\n"
    f.write(data)
f.close()

f = open("F://testlist03.txt", 'w')
for i in range(0, len(test_list_3)):
    if i == len(test_list_3)-1:
        data = test_list_3[i][0] + "/" + test_list_3[i][0] + ".avi"
    else:
        data = test_list_3[i][0] + "/" + test_list_3[i][0] + ".avi" + "\n"
    f.write(data)
f.close()

f = open("F://trainlist01.txt", 'w')
for i in range(0, len(train_list_1)):
    if i == len(train_list_1)-1:
        data = train_list_1[i][0] + "/" + train_list_1[i][0] + ".avi " + train_list_1[i][1]
    else:
        data = train_list_1[i][0] + "/" + train_list_1[i][0] + ".avi " + train_list_1[i][1] + "\n"
    f.write(data)
f.close()

f = open("F://trainlist02.txt", 'w')
for i in range(0, len(train_list_2)):
    if i == len(train_list_2)-1:
        data = train_list_2[i][0] + "/" + train_list_2[i][0] + ".avi " + train_list_2[i][1]
    else:
        data = train_list_2[i][0] + "/" + train_list_2[i][0] + ".avi " + train_list_2[i][1] + "\n"
    f.write(data)
f.close()

f = open("F://trainlist03.txt", 'w')
for i in range(0, len(train_list_3)):
    if i == len(train_list_3)-1:
        data = train_list_3[i][0] + "/" + train_list_3[i][0] + ".avi " + train_list_3[i][1]
    else:
        data = train_list_3[i][0] + "/" + train_list_3[i][0] + ".avi " + train_list_3[i][1] + "\n"
    f.write(data)
f.close()
```
